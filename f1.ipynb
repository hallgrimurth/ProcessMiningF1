```python
# read the telementery data of all laps for Ham and add a column for current lap

Dataframe = pd.DataFrame()
for lap in range(1, 53):
    driver_lap = laps_driver_1.pick_lap(lap)
    telemetry_driver_lap = driver_lap.get_telemetry().add_distance()
    #set Lap as the first column
    telemetry_driver_lap.insert(0, 'CaseID', lap)
    Dataframe = pd.concat([Dataframe, telemetry_driver_lap], axis=0)

#remove the first col
Dataframe.to_csv('silverstone_hamilton_laps.csv')



  
```

```python
#load the df 
data = pd.read_csv('silverstone_hamilton_laps.csv')
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
data['TopSpeedAchieved'] = 0

# Find the index of the row with maximum speed for each CaseID and set 'TopSpeedAchieved' to 1 for those rows
top_speed_indices = data.loc[data.groupby('CaseID')['Speed'].idxmax()].index
data.loc[top_speed_indices, 'TopSpeedAchieved'] = 1

#show the data where top speed is achieved and the speed value  
data[data['TopSpeedAchieved'] == 1][['CaseID', 'Speed']]
```

```python
#save the data
data.to_csv('silverstone_hamilton_laps.csv')
```

```python
#import pandas as pd

# Load the dataset

# Add a 'BrakeApplied' column that is True when the 'Brake' column is True and the previous 'Brake' value was False
data['BrakeApplied'] = (data['Brake'] == True) & (data['Brake'].shift(1) == False)

data['BreakRelease'] = (data['Brake'] == False) & (data['Brake'].shift(1) == True)

data
```
